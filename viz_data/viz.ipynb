```python
# ============================================================
# COMPLETE SCRIPT - ALL CHARTS EXCEPT BOX PLOT
# ============================================================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import numpy as np

# Set up paths
BASE_PATH = r'C:\SQL PROJECTS\DATALAB_INTERN\viz_data'
os.makedirs(BASE_PATH, exist_ok=True)

# Set style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 12

print("=" * 60)
print("STARTING VISUALIZATION...")
print("=" * 60)

# ============================================================
# 1. BAR CHART: Top 10 Countries
# ============================================================
print("\n1. Creating Top Countries Bar Chart...")

df_countries = pd.read_csv(os.path.join(BASE_PATH, 'top_countries.csv'))
df_countries.columns = df_countries.columns.str.strip().str.lower()

plt.figure(figsize=(12, 8))
bars = plt.bar(df_countries['country'], df_countries['transaction_count'], 
               color='steelblue', edgecolor='black', alpha=0.8)

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height):,}', ha='center', va='bottom', fontsize=9)

plt.title('Top 10 Countries by Number of Transactions', fontsize=16, fontweight='bold')
plt.xlabel('Country', fontsize=14)
plt.ylabel('Transaction Count', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig(os.path.join(BASE_PATH, 'chart1_top_countries.png'), dpi=150, bbox_inches='tight')
plt.close()
print("   ✓ Saved: chart1_top_countries.png")

# ============================================================
# 2. BAR CHART: Quantity Distribution
# ============================================================
print("\n2. Creating Quantity Distribution...")

df_quantity = pd.read_csv(os.path.join(BASE_PATH, 'quantity_distribution.csv'))
df_quantity.columns = df_quantity.columns.str.strip().str.lower()

plt.figure(figsize=(12, 8))
bars = plt.bar(df_quantity['quantity'], df_quantity['frequency'], 
               color='coral', edgecolor='black', alpha=0.7)

plt.title('Distribution of Quantity per Transaction', fontsize=16, fontweight='bold')
plt.xlabel('Quantity', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig(os.path.join(BASE_PATH, 'chart2_quantity_distribution.png'), dpi=150, bbox_inches='tight')
plt.close()
print("   ✓ Saved: chart2_quantity_distribution.png")

# ============================================================
# 3. DONUT CHART: Geographic Distribution (Fixed)
# ============================================================
print("\n3. Creating Geographic Distribution Donut Chart...")

df_geo = pd.read_csv(os.path.join(BASE_PATH, 'country_distribution.csv'))
df_geo.columns = df_geo.columns.str.strip().str.lower()

fig, ax = plt.subplots(figsize=(12, 10))

colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0']

wedges, texts, autotexts = ax.pie(
    df_geo['total_transactions'],
    labels=df_geo['country_group'],
    autopct='%1.1f%%',
    colors=colors[:len(df_geo)],
    startangle=90,
    textprops={'fontsize': 13, 'fontweight': 'bold'},
    pctdistance=0.85,
    labeldistance=1.1
)

# Add circle in center to make it a donut
circle = plt.Circle((0, 0), 0.60, fc='white')
ax.add_artist(circle)

for autotext in autotexts:
    autotext.set_color('black')
    autotext.set_fontsize(14)
    autotext.set_fontweight('bold')

plt.title('Geographic Distribution of Transactions', fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()
plt.savefig(os.path.join(BASE_PATH, 'chart3_geo_distribution.png'), dpi=150, bbox_inches='tight')
plt.close()
print("   ✓ Saved: chart3_geo_distribution.png")

# ============================================================
# 4. LINE CHART: Monthly Sales Trend
# ============================================================
print("\n4. Creating Monthly Sales Trend...")

df_monthly = pd.read_csv(os.path.join(BASE_PATH, 'monthly_sales.csv'))
df_monthly.columns = df_monthly.columns.str.strip().str.lower()

df_monthly['month_date'] = pd.to_datetime(df_monthly['month'] + '-01')

fig, ax1 = plt.subplots(figsize=(14, 8))

color = 'steelblue'
ax1.set_xlabel('Month', fontsize=14)
ax1.set_ylabel('Transaction Count', color=color, fontsize=14)
ax1.plot(df_monthly['month_date'], df_monthly['transaction_count'], 
         color=color, marker='o', linewidth=2, label='Transactions')
ax1.tick_params(axis='y', labelcolor=color)
ax1.grid(alpha=0.3)

ax2 = ax1.twinx()
color = 'coral'
ax2.set_ylabel('Total Revenue', color=color, fontsize=14)
ax2.plot(df_monthly['month_date'], df_monthly['total_revenue'], 
         color=color, marker='s', linewidth=2, linestyle='--', label='Revenue')
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Monthly Sales Trend: Transactions and Revenue', fontsize=16, fontweight='bold')
ax1.legend(['Transactions'], loc='upper left')
ax2.legend(['Revenue'], loc='upper right')

plt.tight_layout()
plt.savefig(os.path.join(BASE_PATH, 'chart4_monthly_sales.png'), dpi=150, bbox_inches='tight')
plt.close()
print("   ✓ Saved: chart4_monthly_sales.png")

# ============================================================
# 5. BAR CHART: Top 10 Products
# ============================================================
print("\n5. Creating Top Products Bar Chart...")

df_products = pd.read_csv(os.path.join(BASE_PATH, 'top_products.csv'))
df_products.columns = df_products.columns.str.strip().str.lower()

top_products = df_products.head(10)

plt.figure(figsize=(14, 8))
bars = plt.barh(top_products['description'], top_products['frequency'], 
                color='teal', edgecolor='black', alpha=0.8)

for bar in bars:
    width = bar.get_width()
    plt.text(width, bar.get_y() + bar.get_height()/2.,
             f'{int(width):,}', ha='left', va='center', fontsize=10)

plt.title('Top 10 Most Frequently Purchased Products', fontsize=16, fontweight='bold')
plt.xlabel('Number of Transactions', fontsize=14)
plt.ylabel('Product Description', fontsize=14)
plt.tight_layout()
plt.savefig(os.path.join(BASE_PATH, 'chart5_top_products.png'), dpi=150, bbox_inches='tight')
plt.close()
print("   ✓ Saved: chart5_top_products.png")

# ============================================================
# 6. BAR CHART: Unit Price by Country
# ============================================================
print("\n6. Creating Unit Price by Country Bar Chart...")

df_price = pd.read_csv(os.path.join(BASE_PATH, 'unit_price_by_country.csv'))
df_price.columns = df_price.columns.str.strip().str.lower()
df_price = df_price.sort_values('avg_unit_price', ascending=False).head(10)

plt.figure(figsize=(14, 8))
bars = plt.bar(df_price['country'], df_price['avg_unit_price'], 
               color='skyblue', edgecolor='black', alpha=0.8)

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.05,
             f'{height:.2f}', ha='center', va='bottom', fontsize=11, fontweight='bold')

plt.title('Average Unit Price by Country (Top 10)', fontsize=16, fontweight='bold')
plt.xlabel('Country', fontsize=14)
plt.ylabel('Average Unit Price', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig(os.path.join(BASE_PATH, 'chart6_unit_price_country.png'), dpi=150, bbox_inches='tight')
plt.close()
print("   ✓ Saved: chart6_unit_price_country.png")

# ============================================================
# 7. BAR CHART: Daily Sales Pattern
# ============================================================
print("\n7. Creating Daily Sales Pattern...")

df_daily = pd.read_csv(os.path.join(BASE_PATH, 'daily_sales.csv'))
df_daily.columns = df_daily.columns.str.strip().str.lower()
df_daily['day_of_week'] = df_daily['day_of_week'].str.strip()

day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df_daily['day_order'] = pd.Categorical(df_daily['day_of_week'], categories=day_order, ordered=True)
df_daily = df_daily.sort_values('day_order')

plt.figure(figsize=(12, 8))
bars = plt.bar(df_daily['day_of_week'], df_daily['transaction_count'], 
               color='mediumseagreen', edgecolor='black', alpha=0.8)

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height):,}', ha='center', va='bottom', fontsize=10)

plt.title('Sales by Day of Week', fontsize=16, fontweight='bold')
plt.xlabel('Day of Week', fontsize=14)
plt.ylabel('Transaction Count', fontsize=14)
plt.tight_layout()
plt.savefig(os.path.join(BASE_PATH, 'chart7_daily_sales.png'), dpi=150, bbox_inches='tight')
plt.close()
print("   ✓ Saved: chart7_daily_sales.png")

# ============================================================
# SUMMARY
# ============================================================

print("\n" + "=" * 60)
print("VISUALIZATION COMPLETE!")
print("=" * 60)
print("\nCharts Saved in:", BASE_PATH)
print("\n✓ chart1_top_countries.png - Bar Chart: Top 10 Countries")
print("✓ chart2_quantity_distribution.png - Bar Chart: Quantity Distribution")
print("✓ chart3_geo_distribution.png - Donut Chart: Geographic Distribution")
print("✓ chart4_monthly_sales.png - Line Chart: Monthly Sales Trend")
print("✓ chart5_top_products.png - Bar Chart: Top 10 Products")
print("✓ chart6_unit_price_country.png - Bar Chart: Unit Price by Country")
print("✓ chart7_daily_sales.png - Bar Chart: Daily Sales Pattern")
print("\n" + "=" * 60)
```
